```python
df = spark.read.option("header", True).csv(
    "/Volumes/workspace/default/raw_data",
)

df = df.toDF(*[
    c.replace(" ", "_").replace("-", "_").lower()
    for c in df.columns
])

# Create table
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.retail")

df.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("workspace.retail.bronze_retail_sales")

bronze_df = df
display(bronze_df)
```
