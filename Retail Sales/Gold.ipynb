```python
from pyspark.sql.functions import sum

gold_df = spark.table("workspace.retail.silver_retail_sales")
```

```python
# Sales - Category
spark.sql("DROP TABLE IF EXISTS workspace.retail.gold_sales_daily_category")

gold_sales_daily_category = gold_df.groupBy("date","product_category","gender","customer_id").agg(
    sum("total_amount").alias("revenue"),
    sum("quantity").alias("units_sold"),
)

gold_sales_daily_category.write.format("delta").mode("overwrite").saveAsTable(
    "workspace.retail.gold_sales_daily_category"
)

display(gold_sales_daily_category)
```

```python
# Customers
spark.sql("DROP TABLE IF EXISTS workspace.retail.gold_customers")

gold_customers = gold_df.select("gender", "age", "customer_id","product_category")

gold_customers.write.format("delta").mode("overwrite").saveAsTable(
    "workspace.retail.gold_customers"
)

display(gold_customers)
```

```python
# Sales per Age Group
spark.sql("DROP TABLE IF EXISTS workspace.retail.gold_age_group_sales")

gold_age_group_sales = gold_df.groupBy("customer_id","product_category").agg(
    sum("total_amount").alias("revenue"),
    sum("quantity").alias("units_sold"),
)


gold_age_group_sales.write.format("delta").mode("overwrite").saveAsTable(
    "workspace.retail.gold_age_group_sales"
)

display(gold_age_group_sales)
```
